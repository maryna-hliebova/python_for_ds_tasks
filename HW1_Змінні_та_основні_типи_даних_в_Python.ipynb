**2. Арифметичні операції**

Створіть дві змінні: `var1` зі значенням 15, `var2` зі значенням 4.

Обчисліть та виведіть (кожен пункт оцінюється в 0.5 бала):
- Суму
- Різницю
- Добуток
- Ділення
- Цілочисельне ділення
- Залишок від ділення
- Піднесення до степеня


```python
var1 = 15
var2 = 4

sum1 = var1 + var2
diff = var1 - var2
prod = var1 * var2
div = var1 / var2
int_div = var1 // var2
mod = var1 % var2
exp = var1 ** var2

print("sum is: ", sum1)
print("diff is: ", diff)
print("prod is: ", prod)
print("div is: ", div)
print("int div is: ", int_div)
print("mod is: ", mod)
print("exp is: ", exp)
```
